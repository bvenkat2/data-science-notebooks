# Modifying/Iterating Through Data
There are many different ways to accomplish the same goal in pandas. Some may be better than others. The best performance comes from manipulating the dataframe by columns. Next most ideal is by using the apply function, and then by using iloc or at. The worst case (but sometimes unavoidable) is when using iterrows (or itertuples) to iterate over data. Try to avoid this whenever possible.

```python
for index, row in df.iterrows():
    print(index,row.values)
```

```python
for i in range(len(df)):
    print(i, df.iloc[i].values)
```

```python
def print_rows(row):
    print(row.name,row.values)
df.apply(lambda x: print_rows(x), axis = 1)
```

```python
df.apply(lambda x: print(x.name,x.values), axis = 1)
```

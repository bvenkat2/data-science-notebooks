# Graphing With Matplotlib
One of the first steps in data science is often plotting data to get a sence of how its distributed. This notebook offers an introduction to graphing data with matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline  
```

```python
x = np.linspace(0,100,100)
y = x
plt.plot(x,y)
```

```python
x = np.linspace(-50,50,100)
y = [x_val ** 2 for x_val in x]
plt.plot(x,y)
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(x,y)
```

```python
import random 
x = pd.date_range('2012-10-01', periods=5, freq='5min')
start = 50
y = [start + random.random() * 10 for i in range(5)]
plt.plot(x,y)
```

```python
df = pd.read_csv('Temperature_Data.csv')
```

```python
plt.scatter(df['Temperature'], df['Anomaly'])
plt.title('Temperature Anomalies')
plt.xlabel('Temperature')
plt.ylabel('Anomaly')
```
